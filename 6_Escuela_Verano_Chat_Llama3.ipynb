# Modelos de Masivos de Lenguaje (LLM)

Los Modelos de Lenguaje Grande (LLMs por sus siglas en inglés), también conocidos como modelos de lenguaje de gran tamaño, son un tipo de inteligencia artificial (IA) que se entrenan con grandes cantidades de datos de texto para aprender a generar texto similar al que han sido expuestos.

En otras palabras, los LLMs son como esponjas gigantes para el lenguaje. Absorben grandes cantidades de información textual y luego aprenden a usar esa información para generar nuevo texto que sea similar en estilo, contenido y gramática al texto en el que fueron entrenados.

## ¿Cómo funcionan los LLMs?

Los LLMs funcionan utilizando una arquitectura de red neuronal llamada transformador. Los transformadores son redes neuronales muy potentes que son capaces de procesar y comprender secuencias largas de texto.

A medida que un LLM se entrena, aprende a identificar los patrones y relaciones entre las palabras en el texto. Esta información se utiliza luego para generar nuevo texto que sea similar al texto en el que se entrenó el modelo.


## ¿Qué pueden hacer los LLMs?

Los LLMs tienen una amplia gama de aplicaciones, que incluyen:

* **Generación de texto**: Los LLMs pueden utilizarse para generar texto realista y creativo, como poemas, artículos de noticias, guiones e incluso libros.

* **Traducción de idiomas**: Los LLMs pueden utilizarse para traducir texto de un idioma a otro.

* **Redacción de contenido**: Los LLMs pueden utilizarse para generar contenido de marketing, como descripciones de productos y publicaciones en redes sociales.

* **Resumen de texto**: Los LLMs pueden utilizarse para resumir textos largos en textos más cortos y concisos.

* **Respuesta a preguntas**: Los LLMs pueden utilizarse para responder a preguntas de forma informativa, incluso si son abiertas, desafiantes o extrañas.

* **Detección de sesgos**: Los LLMs pueden utilizarse para detectar sesgos en el texto.

* **Creación de chatbots**: Los LLMs pueden utilizarse para crear chatbots que puedan mantener conversaciones con los humanos.

## ¿Cuáles son algunos ejemplos de LLMs?

* **GPT-3**: Desarrollado por OpenAI, GPT-3 es uno de los LLMs más conocidos y potentes. Es capaz de generar texto realista y creativo, traducir idiomas y responder a preguntas de forma informativa.

* **LLama 3**: Es un modelo de lenguaje grande desarrollado por Meta (anteriormente Facebook). Esta tercera versión continúa la línea de investigación de Meta en la creación de modelos avanzados de procesamiento del lenguaje natural, enfocados en ser eficientes y accesibles para la investigación y aplicaciones prácticas.

* **Mistral**: Es un LLM creado por la startup francesa Mistral AI, fundada en 2023. La empresa se destaca por su enfoque en desarrollar modelos de inteligencia artificial que sean altamente eficientes en términos de rendimiento y capacidad de comprensión del lenguaje natural, con aplicaciones en diversas industrias.

* **Gemini**: Es el LLM desarrollado por Google DeepMind, anteriormente conocido como Google Brain. Gemini se centra en integrar capacidades avanzadas de inteligencia artificial con una fuerte capacidad de comprensión y generación de lenguaje natural, apuntando a ser uno de los modelos más sofisticados y útiles en el mercado de IA.

## ¿Cuáles son las ventajas y desventajas de los LLMs?

**Ventajas**:

* Los LLMs pueden generar texto realista y creativo.

* Los LLMs pueden traducir idiomas con precisión.

* Los LLMs pueden utilizarse para automatizar tareas que requieren comprensión del lenguaje.

* Los LLMs pueden utilizarse para detectar sesgos en el texto.

* Los LLMs pueden utilizarse para crear chatbots que puedan mantener conversaciones con los humanos.

**Desventajas**:

* Los LLMs pueden ser costosos de entrenar y ejecutar.

* Los LLMs pueden ser sesgados, reflejando los sesgos de los datos en los que fueron entrenados.

* Los LLMs pueden generar texto que sea factualmente incorrecto o engañoso.

* Los LLMs pueden ser utilizados para crear deepfakes y otro contenido malicioso.

## ¿En qué se diferencian los LLMs de otros modelos de lenguaje?

Los LLMs se diferencian de otros modelos de lenguaje en los siguientes puntos clave.

* **Tamaño y complejidad**: Los LLMs son modelos de aprendizaje automático mucho más grandes y complejos que los modelos de lenguaje tradicionales. Esto les permite procesar y comprender información de manera más sofisticada, generando resultados más precisos y completos.

* **Datos de entrenamiento**: Los LLMs se entrenan con cantidades masivas de datos de texto, mientras que los modelos de lenguaje tradicionales se entrenan con conjuntos de datos más pequeños y específicos. Esta exposición a una amplia gama de información les da a los LLMs una mejor comprensión del lenguaje y sus matices.

* **Capacidad de generar texto**: Una de las principales diferencias es la capacidad de generar texto. Los LLMs pueden crear texto nuevo, como poemas, código, guiones, piezas musicales, correos electrónicos, cartas, etc., de manera similar a como lo haría un humano. Los modelos de lenguaje tradicionales, por otro lado, generalmente se enfocan en tareas como la traducción o el resumen de texto existente.

* **Comprensión del contexto**: Los LLMs son capaces de comprender el contexto en el que se utiliza el lenguaje, lo que les permite generar respuestas más precisas y relevantes. Los modelos de lenguaje tradicionales a menudo no tienen en cuenta el contexto, lo que puede llevar a respuestas incorrectas o irrelevantes.

* **Habilidades multitarea**: Los LLMs son capaces de realizar una variedad de tareas, como traducción, resumen, escritura creativa y respuesta a preguntas. Los modelos de lenguaje tradicionales generalmente se entrenan para una tarea específica y no funcionan bien en otras áreas.

* **Evolución y aprendizaje continuo**: Los LLMs se pueden actualizar y mejorar continuamente a medida que se exponen a nuevos datos y aprenden de sus interacciones. Los modelos de lenguaje tradicionales generalmente son estáticos y no mejoran con el tiempo.

````python
import pandas as pd
from tqdm.auto import tqdm
import spacy

nlp = spacy.load("es_core_news_sm")

ruta_archivo = "data/Los_Caudillos_de_1830.csv"

# Leer el archivo .csv
df = pd.read_csv(ruta_archivo)

oraciones = []

for texto in tqdm(df['texto'], total=df.shape[0], desc="Obteniendo oraciones"):
    doc = nlp(texto, disable=["ner"])
    for sent in doc.sents:
        tokens = [token.text for token in sent]
        if len(tokens) > 1:
          oraciones.append(sent.text)

print(f"Cantidad de oraciones: {len(oraciones):,}")

# longitud máxima de oraciones
max_len = max(len(oracion) for oracion in oraciones)

print(f"Longitud máxima de oraciones: {max_len}")

````

## Extracción de información con LLMs

Extracción de entidades (nombres de personas, organizaciones, lugares, etc.) y las relaciones que existen entre estas entidades.

### LLama3

En este ejercicio se usa Llama3 cuantizado. [unsloth/llama-3-8b-Instruct-bnb-4bit](https://huggingface.co/unsloth/llama-3-8b-Instruct-bnb-4bit)

**LLM cuantizado**

Un LLM cuantizado es un modelo de lenguaje grande (o masivo) que ha sido optimizado para ser más eficiente en términos de memoria y procesamiento.

La cuantización es una técnica que reduce el tamaño del modelo y la cantidad de recursos necesarios para ejecutarlo, sin perder demasiada precisión. Esto se logra representando los números en el modelo con menos bits (por ejemplo, usando 8 bits en lugar de 32).

En términos simples, es como comprimir una imagen o un archivo para que ocupe menos espacio, pero aún mantenga una calidad aceptable. Esta optimización permite que los LLMs se ejecuten más rápido y en dispositivos con menos capacidad, como teléfonos móviles o computadoras menos potentes.

````python
import transformers
import torch
import pandas as pd

# cargar el modelo Llama3 cuantizado de Hugging Face
model_id = "unsloth/llama-3-8b-Instruct-bnb-4bit"
pipeline = transformers.pipeline(
    "text-generation",
    model=model_id,
    model_kwargs={

        "torch_dtype": torch.float16,
        "quantization_config": {"load_in_4bit": True,},
        "low_cpu_mem_usage": True,
    },
)
````

#### Reconocimiento de entidades (NER)

````python
# definir las instrucciones que se le solicitan al modelo

sentence = "En su afán por defender a los desfavorecidos y luchar contra la injusticia, el ingenioso hidalgo Don Quijote de la Mancha, acompañado por su fiel escudero Sancho Panza, se embarcó en una serie de aventuras junto a su amada Dulcinea del Toboso, enfrentándose a gigantes imaginarios, caballeros andantes como el Caballero de la Blanca Luna y el feroz Basilio, mientras que el sabio Frestón y la astuta Maritornes tejían intrigas y la bella Dorotea interpretaba papeles para ayudar al caballero en sus descabellados planes."

output = '{"entities": [("entidad", "etiqueta")], "relaciones": [("entidad1", "relación", "entidad2")]}'
no_possible = '\{\}'

messages = [
    {"role": "system", "content": "Eres un experto en identificar entidades y las relaciones que existen entre entidades."},
    {"role": "user", "content": f"""
        - Identifca todas las entidades encontradas en la oración.
        - Identifica las relaciones que existen entre las entidades identificadas previamente.
        - El resultado debe estar en formato JSON para cada oración generada con la estructura: {output}.
        - La tarea solicitada se debe aplicar a la oración: ```{sentence}```.
        - Si no puedes identificar entidades, dejar vacía la estructura: {no_possible}".
        - No se requiere información adicional a la proporcionada.
        - Evita comentarios adicionales en la respuesta.
    """},
]

prompt = pipeline.tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=True
)
terminators = [
    pipeline.tokenizer.eos_token_id,
    pipeline.tokenizer.convert_tokens_to_ids("<|eot_id|>")
]
outputs = pipeline(
    prompt,
    max_new_tokens=512,
    eos_token_id=terminators,
    do_sample=True,
    temperature=0.6,
    top_p=0.9,
)
data = outputs[0]["generated_text"][len(prompt):]
print(data)
````

#### NER para Los Caudillos

````python
# definir las instrucciones que se le solicitan al modelo
from tqdm.auto import tqdm

def guardar_datos(ruta, datos, mode='a'):
  with open(ruta, 'w', encoding='utf-8') as f:
    f.write(datos)

# inicializar iniciar/resetear archivo
path_to_save = 'data/resultados_ei.txt'
guardar_datos(path_to_save, '', 'w')

entidades_res = []
relaciones_res = []
for i, sentence in tqdm(enumerate(oraciones), total=len(oraciones), desc="Obteniendo oraciones"):
  output = '{"entities": [("entidad", "etiqueta")], "relaciones": [("entidad1", "relación", "entidad2")]}'
  no_possible = '\{\}'

  messages = [
      {"role": "system", "content": "Eres un experto en identificar entidades y las relaciones que existen entre entidades."},
      {"role": "user", "content": f"""
          - Identifca todas las entidades encontradas en la oración.
          - Identifica las relaciones que existen entre las entidades identificadas previamente.
          - El resultado debe estar en formato JSON para cada oración generada con la estructura: {output}.
          - La tarea solicitada se debe aplicar a la oración: ```{sentence}```.
          - Si no puedes identificar entidades, dejar vacía la estructura: {no_possible}".
          - No se requiere información adicional a la proporcionada.
          - Evita comentarios adicionales en la respuesta.
      """},
  ]

  prompt = pipeline.tokenizer.apply_chat_template(
          messages,
          tokenize=False,
          add_generation_prompt=True
  )
  terminators = [
      pipeline.tokenizer.eos_token_id,
      pipeline.tokenizer.convert_tokens_to_ids("<|eot_id|>")
  ]
  outputs = pipeline(
      prompt,
      max_new_tokens=512,
      eos_token_id=terminators,
      do_sample=True,
      temperature=0.6,
      top_p=0.9,
  )
  data = outputs[0]["generated_text"][len(prompt):]
  # procesar e imprimir solo las primeras 20 líneas
  if i > 20:
    print(f"{i+1}\t{data}")
    break
  #guardar_datos(path_to_save, f'{data}\n')

  # almacenar la salida en una estructura
  if type(data) == str:
    try:
      data = eval(data)
      entidades_res.append(data['entities'])
      if len(data['relaciones']) > 0:
        relaciones_res.append(data['relaciones'])
      print('-'*50)
      print("El resultado es una cadena.")
      print(data)
    except Exception as e:
      print(f"Error al procesar la salida: {e}\n\n")
      print(f"{data}\n\n")
  elif type(data) == dict:
    entidades_res.append(data['entities'])
    if len(data['relaciones']) > 0:
        relaciones_res.append(data['relaciones'])
    print('-'*50)
    print("El resultado es una estructura.")
    print(data)
````

````python
# crear un DataFrame con las entiddades y relaciones obtenidas
tmp_entidades = {
    "entity": [],
    "label": [],
}
for sub_list in entidades_res:
  for entidad, etiqueta in sub_list:
    tmp_entidades["entity"].append(entidad)
    tmp_entidades["label"].append(etiqueta.upper())

df_entidades = pd.DataFrame(tmp_entidades)
````

````python
tmp_relaciones = {
    "entity1": [],
    "relation": [],
    "entity2": [],
}

for sub_list in relaciones_res:
  for item in sub_list:
    tmp_relaciones["entity1"].append(item[0].lower())
    tmp_relaciones["relation"].append(item[1].lower())
    tmp_relaciones["entity2"].append(item[2].lower())

df_relaciones = pd.DataFrame(tmp_relaciones)
````

#### Visualizar relaciones

````python
# Instalar las bibliotecas necesarias
# !pip install networkx matplotlib

import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

# Crear un grafo dirigido
G = nx.DiGraph()

# Añadir nodos y relaciones al grafo
for _, row in df_relaciones.iterrows():
    G.add_edge(row['entity1'], row['entity2'], relation=row['relation'])

# Imprimir nodos y aristas del grafo
print("Nodos del grafo:")
print(G.nodes)
print("\nAristas del grafo:")
print(G.edges(data=True))

# Dibujar el grafo
plt.figure(figsize=(10, 8))
pos = nx.spring_layout(G)

# Dibujar nodos y aristas
nx.draw(G, pos, with_labels=True, node_size=3000, node_color="skyblue", font_size=10, font_color="black", font_weight="bold", edge_color="gray")

# Añadir etiquetas a las aristas
edge_labels = nx.get_edge_attributes(G, 'relation')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='red')

plt.title("Visualización del Grafo")
plt.show()
````
